```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

df = pd.read_csv('mahasiswa_data.csv')

df['JENIS KELAMIN'] = df['JENIS KELAMIN'].map({'LAKI - LAKI': 0, 'PEREMPUAN': 1})
df['STATUS PERNIKAHAN'] = df['STATUS PERNIKAHAN'].map({'BELUM': 0, 'MENIKAH': 1})
df['STATUS MAHASISWA'] = df['STATUS MAHASISWA'].map({'MAHASISWA': 0, 'BEKERJA': 1})
df['IPK SEMESTER 1-6'] = df['IPK SEMESTER 1-6'].astype(float)
df['STATUS KELULUSAN'] = df['STATUS KELULUSAN'].map({'TERLAMBAT': 0, 'TEPAT': 1})

X = df[['JENIS KELAMIN', 'STATUS MAHASISWA', 'STATUS PERNIKAHAN', 'IPK SEMESTER 1-6', 'STATUS KELULUSAN']]
y = df['STATUS KELULUSAN']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = LogisticRegression()
model.fit(X_train, y_train)

df['PREDIKSI STATUS KELULUSAN'] = model.predict(X)

df
```

```python
# confusion matrix
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
